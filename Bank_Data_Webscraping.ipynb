```python
#  Keyword Groups
keywords = [
    # General & Brand Awareness
    "best bank in nigeria", "top banks in nigeria", "best bank for savings in nigeria",
    "open bank account online nigeria", "online banking nigeria",
"nigerian bank interest rates",
    "bank charges in nigeria", "customer service in nigerian banks",
    # Savings & Account Management
    "open savings account nigeria", "best savings account nigeria",
"savings account interest rate nigeria",
    "fixed deposit account nigeria", "current account nigeria",
"domiciliary account nigeria",
    "student savings account nigeria", "business account nigeria",
"minimum balance for savings account nigeria",
    # Loans & Credit
    "personal loan nigeria", "business loan nigeria", "salary loan nigeria", "instant loan nigeria",
    "student loan nigeria", "car loan nigeria", "mortgage loan nigeria", "agriculture loan nigeria",
    "small business loan nigeria", "bank loan requirements nigeria",
    # Digital / Internet Banking
    "internet banking nigeria", "mobile banking nigeria", "ussd code for bank transfer nigeria",
    "online money transfer nigeria", "best online banking app nigeria",
    "how to activate internet banking nigeria", "mobile banking app download nigeria",
    "how to reset internet banking password nigeria", "how to register for mobile banking nigeria",
    #  Cards & Payments
    "atm card nigeria", "debit card nigeria", "credit card nigeria",
"virtual card nigeria",
    "prepaid card nigeria", "cardless withdrawal nigeria",
"international atm card nigeria",
    #  Investment & Financial Products
    "bank investment plans nigeria", "fixed income investment nigeria", "treasury bills nigeria",
    "wealth management nigeria", "mutual funds nigeria", "bank insurance products nigeria",
    "retirement savings account nigeria"
]
# Bank Names to Track
banks = [
    "GTBank", "GT Bank", "Access Bank", "United Bank for Africa","UBA", "Zenith Bank",
"Fidelity Bank",
    "Union Bank", "First Bank", "Sterling Bank", "Wema Bank",
"Keystone Bank",
    "Stanbic IBTC", "Polaris Bank", "Unity Bank", "Heritage Bank",
"Providus Bank", "Opay", "Moniepoint", "Kuda", "Ecobank","Jaiz Bank","Parallex Bank", "Globus Bank"
]
```

```python
# Data Collection
results_list = []
for keyword in keywords:
    print(f"Fetching SERP data for: {keyword} ...")
    params = {
        "q": keyword,
        "location": "Nigeria",
        "google_domain": "google.com.ng",
        "hl": "en",
        "gl": "ng",
        "api_key": api_key
    }
    try:
        search = GoogleSearch(params)
        results = search.get_dict()
        if "organic_results" in results:
            for idx, res in enumerate(results["organic_results"],
start=1):
                title = res.get("title", "")
                link = res.get("link", "")
                matched_banks = [
                    b for b in banks
                    if b.lower().replace(" ", "") in
title.lower().replace(" ", "")
                    or b.lower().replace(" ", "") in
link.lower().replace(" ", "")
                ]
                for bank in matched_banks:
                    results_list.append({
                        "keyword": keyword,
                        "rank_position": idx,
                        "title": title,
                        "link": link,
                        "bank": bank
                    })
        else:
            print(" No organic results found for:", keyword)
    except Exception as e:
        print(f"Error fetching {keyword}: {e}")
    # Sleep to avoid hitting rate limits
    time.sleep(2)
```

```python
# Create DataFrame
df = pd.DataFrame(results_list)
print(" Total results collected:", len(df))
df.head()
```

```python
# Clean and save dataset
df.drop_duplicates(inplace=True)
df.dropna(subset=['bank'], inplace=True)
# Add a visibility score
df['visibility_score'] = 1 / df['rank_position']
# Save as CSV for Tableau
df.to_csv("nigerian_banks_seo_dataset.csv", index=False)
print("Dataset saved successfully: nigerian_banks_seo_dataset.csv")
df.sample(10)
```
